```python
def create_tf_example(features, label):
    # Examples maps string keys to either ints, floats or bytes
    tf_example = tf.train.Example(features=tf.train.Features(feature={
        'idx': tf.train.Feature(int64_list=tf.train.Int64List(value=[features[0]])),
        'text': tf.train.Feature(bytes_list=tf.train.BytesList(value=[features[1].encode('utf-8')])),
        # Label is the sentiment value we are trying to predict
        'label': tf.train.Feature(int64_list=tf.train.Int64List(value=[label]))
    }))

    # Encoded example can be extracted strings, 
    return tf_example
```

```python
from tensorflow.python.lib.io.tf_record import TFRecordWriter

def convert_csv_to_tfrecord(csv, file_name):
    writer = TFRecordWriter(file_name)
    for row in csv:
        # features are the predictive values. Label is the value you are trying to predict.
        features, label = row[:-1], row[-1]
        # print(features)
        example = create_tf_example(features, label)
        writer.write(example.SerializeToString())
    writer.close()
```

```python
def decode(serialized_example):
    """
    Parses an image and label from the given `serialized_example`.
    It is used as a map function for `dataset.map`
    """

    features = tf.compat.v1.parse_single_example(
        serialized_example,
        features={
        'idx': tf.io.FixedLenFeature([], tf.int64),
        'text': tf.io.FixedLenFeature([], tf.string),
        'label': tf.io.FixedLenFeature([], tf.int64)
           })
    
    
    
    text = tf.cast(features['text'], tf.string)
    label = tf.cast(features['label'], tf.int64)
    idx = tf.cast(features['idx'], tf.int64)

    return text, label
```

```python


def clean_string(features):
    revised_sentence = tf.strings.regex_replace(features['text'], "\.\.\.", "", replace_global=True)
    revised_sentence = tf.strings.regex_replace(revised_sentence, "\\'", "'", replace_global=True)
    revised_sentence = tf.strings.regex_replace(revised_sentence, "\\n", "", replace_global=True)
    features['sentence'] = revised_sentence
    return features

```
